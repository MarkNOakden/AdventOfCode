```python
import pandas as pd
```

```python
triples = pd.read_table('input.txt', header=None, delim_whitespace=True)
```

```python
def valid(a,b,c):
    A, B, C = sorted([a, b, c])
    return not ((A + B) <= C)
```

```python
triples['isvalid'] = triples.apply(lambda row: valid(row[0], row[1], row[2]), axis=1)
```
