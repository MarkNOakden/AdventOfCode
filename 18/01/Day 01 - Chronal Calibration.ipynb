**`--- Day 1: Chronal Calibration ---`**

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_series = pd.read_table('input.txt', header=None, names=['changes'], squeeze=True)
```

## Part 1

```python
startval = 0
initial = pd.Series([startval])
changes = pd.concat([initial, input_series], ignore_index=True)

total = changes.cumsum()
print(f'Part 1: {total.iloc[-1]}')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,4), squeeze=True)
ax1.plot(total[0:578])
ax1.set_title('Prior to big upward jump')
ax2.plot(total)
ax2.set_title('All')
ax3.plot(total[578:993])
ax3.set_title('Between the big jumps')
plt.show()
```

## Part 2

```python
seen = set(total)
cumulative = total.iloc[-1]
finished = False
rounds = 0
iters = len(input_series)

while not finished:
    rounds += 1
    for change in input_series:
        iters += 1
        cumulative += change
        if cumulative in seen:
            print(f'Part 2: {cumulative}')
            print(f'  (After {iters} shifts ({rounds} rounds))')
            finished = True
            break
        else:
            seen.add(cumulative) 
```
