````python
fw['part1pos'] = fw['depth'] % (2 * fw['range'] - 2)
````

````python
fw['caught'] = (fw['part1pos'] == 0)
fw['severity'] = fw['caught'] * fw['depth'] * fw['range']
print(f'part1 answer: {fw.severity.sum()}')
````

````python
fw['modulus'] = (2 * fw['range'] - 2)
fw = fw.sort_values('modulus')
````

````python
%%timeit
delay = 3960000
while True:
    sol = True
    for i in fw[['depth', 'modulus']].values:
        if (i[0] + delay) % i[1] == 0:
            sol = False
            break
    if sol:
        print(f'part 2 answer {delay}')
        break       
    if delay % 10000 == 0:
        print(f'{delay}')
    delay += 1
    
````
